```python
#Librerias
import pandas as pd
import numpy as np

# 1.1 Cargar ventas.csv en carpeta donde está el codigo
df_ventas = pd.read_csv("ventas.csv")

# 1.2 Mostrar archivo (primeras 5 lineas)
print(df_ventas.head(5))

# 2.1 identificar valores nulos
print("\n2. Columnas con nulos y su cantidad:" )
col_nul = df_ventas.isnull().sum()[df_ventas.isnull().sum() > 0]
print(col_nul)
```

```python
df_ventas["Categoría"] = df_ventas["Categoría"].fillna("Desconocido") # Rellenar con texto
df_ventas["Precio"] = df_ventas["Precio"].fillna(df_ventas["Precio"].mean()) # Rellenar con promedio


print(df_ventas)

# 3. Eliminar duplicados
print("\nN° Duplicados:", df_ventas.duplicated().sum())
df_ventas = df_ventas.drop_duplicates()

# 4. Detección y manejo de outliers en Cantidad
q1 = df_ventas["Cantidad"].quantile(0.25)
q3 = df_ventas["Cantidad"].quantile(0.75)
iqr = q3 - q1
rango_1 = q1 - 1.5 * iqr
rango_2 = q3 + 1.5 * iqr
# Filtrar dentro de rango permitido
df_ventas = df_ventas[(df_ventas["Cantidad"] >= rango_1) & (df_ventas["Cantidad"] <= rango_2)]


# 5. Reestructuracion y Modificación e estructura
# Convertir en formato Fecha
df_ventas["Fecha"] = pd.to_datetime(df_ventas["Fecha"], errors="coerce")
# Estandarizar valores tipo texto 
columnas_texto = df_ventas.select_dtypes(include="object").columns
for col in columnas_texto:
    df_ventas[col] = df_ventas[col].astype(str).str.strip().str.title()

# Agregar columna Total
df_ventas["Total"] = df_ventas["Cantidad"] * df_ventas["Precio"]
# Clasificar tipo de cliente
df_ventas["Tipo_Cliente"] = df_ventas["Total"].apply(lambda x: "Preferencial" if x > 200 else "Básico")
# Cambiar indice
df_ventas.set_index("Cliente_ID", inplace=True)
print(df_ventas)
```
