```python
# 1.1 Importar libreria
import pandas as pd

# 1.2 Crear Dataframe 
datos_jugadores = {
    "Jugador": ["Lionel Messi", "Cristiano Ronaldo", "Kevin De Bruyne", "Kylian Mbappé", "Luka Modric"],
    "Posición": ["Delantero", "Delantero", "Mediocampista", "Delantero", "Mediocampista"],
    "Edad": [35, 38, 31, 24, 37],
    "Goles": [20, 18, 8, 25, 3],
    "Asistencias": [10, 5, 15, 12, 8]
}

df_datos_jugadores = pd.DataFrame(datos_jugadores)

# 2. Seleccion de columna y mostrar nombres de jugadores
nombres_jugadores = df_datos_jugadores["Jugador"]
print("\nNombres Jugadores:\n")
print(nombres_jugadores)

# 3. Filtro jugadores con >10 goles
filtro_mas_10_goles = df_datos_jugadores[df_datos_jugadores["Goles"] > 10][["Jugador", "Goles"]]
print("\n",filtro_mas_10_goles)

# 4. Se agrega columna Puntos
df_datos_jugadores["Puntos"] = (df_datos_jugadores["Goles"] * 4) + (df_datos_jugadores["Asistencias"] * 2)

print(f"\nSe agrega columna Puntos:\n {df_datos_jugadores}")

# 5. Promedio de goles
promedio_goles = df_datos_jugadores["Goles"].mean()
print(f"\nPromedio de goles: {promedio_goles}\n")

# 6. Máximo y mínimo de asistencias
max_asistencia = df_datos_jugadores["Asistencias"].max()
min_asistencias = df_datos_jugadores["Asistencias"].min()
print(f"\nMáximo de asistencias: {max_asistencia}")
print(f"Mínimo de asistencias: {min_asistencias}\n")

# 7 y 10. Jugadores por posición
jugadores_posicion = df_datos_jugadores["Posición"].value_counts()
print("\nJugadores por posición:\n")
print(jugadores_posicion)

# 8. Ordenar goles de forma descendente
df_goles_ordenados = df_datos_jugadores.sort_values(by="Goles", ascending=False)
print("\n\nDataFrame ordenado por goles:\n")
print(df_goles_ordenados)

# 9. Estadísticas del df
print("\nEstadísticas de la tabla:\n")
print(df_datos_jugadores.describe())

```
