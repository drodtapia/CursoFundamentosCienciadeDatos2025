```python
# 1. Definir la lista de libros
libros = [
    {"titulo": "Sin familia", "autor": "Hector Malot", "precio": float(10390), "stock": int(17)},
    {"titulo": "Papaito Piernas Largas", "autor": "Jean Webster", "precio": float(9171), "stock": int(22)},
    {"titulo": "Trece casos misteriosos", "autor": "Ana María Güiraldes y Jacqueline Balcells", "precio": float(9900), "stock": int(3)},
    {"titulo": "El lado oculto de la ciencia", "autor": "Fernando Izaurieta", "precio": float(18900), "stock": int(13)},
    {"titulo": "Vida artificial", "autor": "Eric Goles", "precio": float(19900), "stock": int(4)}
]

# 2. Mostrar libros disponibles
def mostrar_libros_disponibles():
    print("\nLibros disponibles (más de 1 en stock):")
    for libro in libros:
        if libro["stock"] > 1:
            print(f"Título: {libro['titulo']}, Autor: {libro['autor']}, Precio: {libro['precio']}, Stock: {libro['stock']}")

# 3. Filtrar libros por rango de precios
def rango_precio():
    minimo = float(input("Ingrese el precio mínimo: "))
    maximo = float(input("Ingrese el precio máximo: "))

    if minimo < maximo:
        libros_encontrados = False
        for libro in libros:
            if minimo <= libro["precio"] <= maximo:
                print(f"Título: {libro['titulo']}, Autor: {libro['autor']}, Precio: {libro['precio']}, Stock: {libro['stock']}")
                libros_encontrados = True
        if not libros_encontrados:
            print("No se encontraron libros en el rango de precios indicado.")
    elif minimo == maximo:
        print("El precio mínimo y máximo son iguales. Ingrese un rango válido.")
    else:
        print("Error: El precio mínimo no puede ser mayor que el máximo.")


# 6. Descuentos por autor
descuentos_autor = {
    "Hector Malot": 0.15  
}

# 4. Comprar libros
def comprar_libros(titulo, cantidad):
    for libro in libros:
        if libro["titulo"].lower() == titulo.lower(): #nNormalizar Titulo a minusculas
            #Verificacion si está el libro en Stock
            if cantidad <= libro["stock"]:
                descuento = descuentos_autor.get(libro["autor"], 0)   #descuento por autor si corresponde sino 0
                monto_final = libro["precio"] * cantidad * (1 - descuento) #En el caso de que tenga descuento se aplica
                ahorro = libro["precio"] * cantidad * descuento
                libro["stock"] -= cantidad
                print(f"Compra exitosa: {cantidad} x {libro['titulo']} - Total: ${monto_final} CLP")
                return monto_final, ahorro
            else:
                print("Error: La cantidad solicitada es mayor al stock disponible.")
                return 0, 0
    print("Error: Libro no encontrado.")
    return 0, 0

# Variables para la factura final
total_pagado = 0
total_ahorro = 0
total_libros_comprados = 0

# 5. Bucle while para realizar múltiples compras
while True:
    print("\n--- Sistema de Compras ---")
    print("1. Mostrar libros disponibles")
    print("2. Filtrar libros por rango de precios")
    print("3. Comprar libro")
    print("4. Finalizar compra y mostrar factura")
    opcion = input("Seleccione una opción (o escriba 'Salir' o presione Enter para salir): ").strip()

    if opcion.lower() == "salir" or opcion == "" or opcion == "4":
        print("\n--- Factura Final ---")
        print(f"Total de libros comprados: {total_libros_comprados}")
        print(f"Monto total pagado: ${total_pagado} CLP")
        print(f"Descuentos aplicados: ${total_ahorro} CLP")
        print("¡Gracias por su compra!")
        break

    if opcion == "1":
        print("\n")
        mostrar_libros_disponibles()
    elif opcion == "2":
        print("\n")
        rango_precio()
    elif opcion == "3":
        print("\n")
        titulo = input("Ingrese el título del libro a comprar: ")
        cantidad = int(input("Ingrese la cantidad deseada: "))
        monto, ahorro = comprar_libros(titulo, cantidad)
        if monto > 0:
            total_pagado += monto
            total_ahorro += ahorro
            total_libros_comprados += cantidad
    else:
        print("Opción no válida. Intente de nuevo.")

```
